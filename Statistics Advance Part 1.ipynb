### 1. **What is a random variable in probability theory?**

 A random variable is a numerical outcome of a random process or experiment. It maps outcomes of a sample space to real numbers.



### 2. **What are the types of random variables?**

* Discrete Random Variable.
* Continuous Random Variable.



### 3. **What is the difference between discrete and continuous distributions?**

* **Discrete distributions** represent outcomes of discrete random variables (e.g., binomial, Poisson).
* **Continuous distributions** represent outcomes of continuous variables (e.g., normal, exponential).



### 4. **What are probability distribution functions (PDF)?**

PDFs describe the likelihood of a random variable to take a specific value. For continuous variables, it represents the only density at a point.



### 5. **How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**

* **PDF** gives the relative likelihood of a value.
* **CDF** gives the probability that a variable is less than or equal to a certain value.



### 6. **What is a discrete uniform distribution?**

A distribution where each possible discrete outcome has an equal probability. Example: Rolling a fair die.



### 7. **What are the key properties of a Bernoulli distribution?**

* Two possible outcomes: success (1) or failure (0).
* Defined by a single parameter $p$, the probability of success.
* Mean = $p$, Variance = $p(1 - p)$.



### 8. **What is the binomial distribution, and how is it used in probability?**

It models the number of successes in $n$ independent Bernoulli trials with probability $p$ of success. Used in quality control, genetics, etc.



### 9. **What is the Poisson distribution and where is it applied?**

Models the number of times an event occurs in a fixed interval (time/space), given a known average rate $\lambda$. Common in queuing theory and traffic analysis.



### 10. **What is a continuous uniform distribution?**

All outcomes in a continuous range have equal probability. Example: Random number generator in \[0,1].



### 11. **What are the characteristics of a normal distribution?**

* Bell-shaped, symmetric about the mean.
* Defined by mean $\mu$ and standard deviation $\sigma$.
* Follows empirical rule: 68%-95%-99.7%.



### 12. **What is the standard normal distribution, and why is it important?**

A normal distribution with $\mu = 0$ and $\sigma = 1$. It's the reference for calculating Z-scores and performing hypothesis tests.



### 13. **What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

It states that the sampling distribution of the sample mean approaches a normal distribution as sample size increases, regardless of the population's distribution. Critical for making inferences.



### 14. **How does the Central Limit Theorem relate to the normal distribution?**

It explains why normal distribution is used in practice: sample means tend to be normally distributed due to the CLT, even if the population is not.



### 15. **What is the application of Z statistics in hypothesis testing?**

Z-statistics help determine if a sample mean significantly differs from a known population mean, especially when population variance is known.



### 16. **How do you calculate a Z-score, and what does it represent?**

$$
Z = \frac{(X - \mu)}{\sigma}
$$

It represents how many standard deviations an observation $X$ is from the mean $\mu$.



### 17. **What are point estimates and interval estimates in statistics?**

* **Point Estimate**: Single value estimate (e.g., sample mean).
* **Interval Estimate**: Range likely to contain the population parameter (e.g., confidence interval).



### 18. **What is the significance of confidence intervals in statistical analysis?**

They provide a range of values for the population parameter and help quantify uncertainty in estimates.



### 19. **What is the relationship between a Z-score and a confidence interval?**

Confidence intervals are calculated using Z-scores for known confidence levels (e.g., Z = 1.96 for 95%).



### 20. **How are Z-scores used to compare different distributions?**

Z-scores standardize values from different distributions, allowing comparison across scales by measuring relative positions.



### 21. **What are the assumptions for applying the Central Limit Theorem?**

* Independent observations
* Identically distributed samples
* Finite variance
* Large enough sample size (typically $n \geq 30$)



### 22. **What is the concept of expected value in a probability distribution?**

It is the weighted average of all possible values a random variable can take:

$$
E(X) = \sum x \cdot P(x) \text{ (discrete)} \quad \text{or} \quad E(X) = \int x f(x) dx \text{ (continuous)}
$$



### 23. **How does a probability distribution relate to the expected outcome of a random variable?**

The expected value is derived from the probability distribution and represents the long-run average outcome of the random variable.
